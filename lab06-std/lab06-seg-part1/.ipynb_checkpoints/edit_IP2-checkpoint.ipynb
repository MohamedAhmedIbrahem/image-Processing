```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
from skimage.color import rgb2gray, gray2rgb
from skimage import data
from scipy.ndimage import gaussian_filter
from skimage.filters import threshold_otsu, threshold_yen, threshold_minimum,threshold_local,threshold_triangle
from sklearn import cluster
from sklearn.cluster import KMeans, DBSCAN


import math 

from commonfunctions import *
%matplotlib inline
%load_ext autoreload
%autoreload 2

from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
```

```python
image = io.imread('4-Figure2.png')
image = rgb2gray(image)
simage = gaussian_filter(image, sigma=0.8)
#edge_sobel = prewitt(simage)
#edge_sobel = edge_sobel < threshold_otsu(edge_sobel)

show_images([image,simage])
```

```python
alpha = 1.2
P =[-3,-2,-1,0,1,2,3]
Q =[-3,-2,-1,0,1,2,3]
k = [2,3,4,5,6,10,11,12,13,14]
size_of_kernal=7
list_kernals=[]
arr_orient = []
edges_per_orient ={}
for i in range(10):
    kernal=np.zeros((7,7))
    angle = k[i]*math.degrees(22.5)
    arr_orient.append(angle)
    edges_per_orient[angle]=0
    #angle = angle*(math.pi/180)
    sumk=0
    for p in range(size_of_kernal):
        for q in range(size_of_kernal):  
            sumGNeg=(1/(2*np.pi*pow(alpha,2))) * np.exp(- (pow(P[p] - (alpha*np.cos(angle)),2) + pow(Q[q]- (alpha*np.sin(angle)),2))/(2*pow(alpha,2)))
            sumGPos=(1/(2*np.pi*pow(alpha,2))) * np.exp(- (pow(P[p] + (alpha*np.cos(angle)),2) + pow(Q[q]+ (alpha*np.sin(angle)),2))/(2*pow(alpha,2)))
            kernal[p,q]=sumGNeg-sumGPos
            if kernal[p,q]>0:
                sumk+=kernal[p,q]
    kernal = kernal / sumk                    
    list_kernals.append(kernal)

arr =[]
for i in range(10):
    CImg = convolve2d(simage,list_kernals[i])
    arr.append(CImg)

out = np.zeros(simage.shape)
Orient = np.zeros(simage.shape)
listR = []
for i in range(simage.shape[0]):
    for j in range(simage.shape[1]):
        for KE in range (10):
            listR.append(arr[KE][i,j])
        out[i,j] = np.amax(listR)
        Orient[i,j] = np.where(listR == out[i,j])[0]
        listR.clear()

ther = np.average(simage)

#out = out < pow(threshold_otsu(simage),2)
#out = out < threshold_otsu(out)/2
#adaptive_thresh = threshold_local(out, 55)
#out = out < adaptive_thresh*2
threshold = threshold_triangle(out)
out = 1 * (out < threshold)
#yen_threshold = threshold_yen(out)
#bright = rescale_intensity(out, (0, yen_threshold), (0, 255))
show_images([simage,out])
```

```python

img_v = out.reshape(out.shape[0]*out.shape[1],1)
#kmeans = cluster.KMeans(n_clusters=7).fit(img_v)
clustring = DBSCAN(eps=0.3, min_samples=2).fit(img_v)
dic = {0:50 , 1 : 150, 2 : 200, 2: 250}
out_img = np.zeros((out.shape[0],out.shape[1]))
k = 0
for i in range(out_img.shape[0]):
    for j in range(out_img.shape[1]):
        out_img[i,j] = dic[clustring.labels_[k]]
        k += 1
        
show_images([out, out_img ])        
```
