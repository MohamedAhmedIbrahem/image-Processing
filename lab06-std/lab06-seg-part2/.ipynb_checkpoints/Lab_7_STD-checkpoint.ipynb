```python
img = io.imread('book.png')
img = rgb2gray(img)
def segment(img):
    img.astype(np.uint8)
    hist = histogram(img)
    hist = hist[0]
    initial_w=0
    size = img.shape[0]*img.shape[1]
    def get_w(hist,begin,end,size):
        w=0
        begin=int(begin)
        end=int(end)
        for i in range(begin,end):
            w+=(i*hist[i]/size)
        return np.round(w)
    Tin = get_w(hist,0,len(hist),size)
    Tout=0
    while(np.abs(Tin-Tout)>2):
        Tout=Tin
        T1=get_w(hist,0,Tin,size)
        T2=get_w(hist,Tin,len(hist),size)
        Tin=np.round((T1+T2)/2)
    binary = (img>=Tin)
    return binary
show_images([img,segment(img)],["original","segmented"])
```

```python
img = io.imread('book.png')
img = rgb2gray(img)
midx=int(img.shape[0]/2)
midy=int(img.shape[1]/2)
img1,img2,img3,img4=np.copy(img[0:midx+1,0:midy+1]),np.copy(img[midx+1:img.shape[0],midy+1:img.shape[1]]),np.copy(img[midx+1:img.shape[0],0:midy+1]),np.copy(img[0:midx+1,midy+1:img.shape[1]])
bin1,bin2,bin3,bin4=segment(img1),segment(img2),segment(img3),segment(img4)
output_img=np.zeros(img.shape)
output_img[0:midx+1,0:midy+1]=np.copy(bin1)
output_img[midx+1:img.shape[0],midy+1:img.shape[1]]=np.copy(bin2)
output_img[midx+1:img.shape[0],0:midy+1]=np.copy(bin3)
output_img[0:midx+1,midy+1:img.shape[1]]=np.copy(bin4)
show_images([img,segment(img),output_img],["original","segmented1","segmented2"])
```

```python
img = io.imread('book.png')
img = rgb2gray(img)
def segment_into_n(img,n):
    stepx = int(img.shape[0]/n)
    stepy = int(img.shape[1]/n)
    x = [0]
    y = [0]
    for i in range(n-1):
        x.append(x[len(x)-1]+stepx)
        y.append(y[len(y)-1]+stepy)
    x.append(img.shape[0])
    y.append(img.shape[1])
    output_img=np.zeros(img.shape)
    for i in range(1,len(x)):
        for j in range(1,len(y)):
            tmp_img=np.copy(img[x[i-1]:x[i],y[j-1]:y[j]])
            output_img[x[i-1]:x[i],y[j-1]:y[j]]=np.copy(segment(tmp_img))
    show_images([img,segment(img),output_img],["original","segmented1","segmented2"])
segment_into_n(img,16)
```
