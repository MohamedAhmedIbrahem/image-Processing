```python
image = io.imread('balls.jpg')

[x, y, z ]= image.shape
img2d = image.reshape(x*y, z)

kmeans_cluster = cluster.KMeans(n_clusters=7)
kmeans_cluster.fit(img2d)
cluster_centers = kmeans_cluster.cluster_centers_
print('cluster_centers')
print(cluster_centers)
print(cluster_centers.shape)
cluster_labels = kmeans_cluster.labels_
print('cluster_labels')
print(cluster_labels.shape)
print(cluster_centers[cluster_labels].shape)

'''
cluster_centers: contains the colors of the new segments
cluster_labels contains the labes for each pixel in the image(don't forget that the image is now converted 
to vector)

Use the same concept shown above

'''
clusteredImg  =cluster_centers[cluster_labels].reshape(x, y, z)
clusteredImg  = clusteredImg.astype('uint8')
show_images([image,clusteredImg])
```

```python
'''
This experiment should be done on Gray scale and on one channel of RGB or HSV
'''

images =['amazing.jpg','balls.jpg','circuit.tif','colors.jpg']
for img in images:
    img = io.imread(img)
    orig = np.copy(img)
    channel = rgb2gray(img)
    thresh = threshold_otsu(channel)
    
    mask = channel > thresh
    if(len(orig.shape)==3):
        img[mask == True] = [255,255,255]
        img[mask == False] = [0,0,0]
    else:
        img[mask == True] = 255
        img[mask == False] = 0
    show_images([orig,img])
```
